# data normalization steps
## column 'data'

```python
dict(data1['data'])
pd.DataFrame(dict(data1['data']))
data1_datacol = pd.DataFrame(dict(data1['data'])).transpose()
data1_datacol.head()

# now I can see what's going on in the column 'data' because I converted it to a dataframe
```

## columns 'purchaser', 'type'

```python
def flatten(data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
```

```python
nested_columns = ['purchaser', 'type']

flat_data1_datacol = flatten(data1_datacol, nested_columns)
flat_data1_datacol.head()

# columns 'purchaser' and 'type' are flattened too
# now I can deal with the nested columns in data1_datacol
```

## column 'awarded'

```python
# column 'awarded' has a very nested dictionary list structure so decided to deal with a function
```

```python
flat_data1_datacol['awarded'][0][0]
```

```python
flat_data1_datacol['awarded'][1][0]
```

```python
dict_new = dict()
for i in range(100):
    awardedcol=flat_data1_datacol['awarded'][i][0]
    for key, val in awardedcol.items():
        if key != "suppliers" and key != "offers_count_data" and key != "offers_count_status":            
            if key not in dict_new:
                dict_new[key] = [val]
            else:
                dict_new[key].append(val)
```

```python
dict_new
```

```python
for key in dict_new.keys():
    print (key)
    print (len(dict_new[key]))
```

```python
flattened_col_awarded = pd.DataFrame(dict_new)
#     print(pd.DataFrame(dict(flat_data1_datacol['awarded'])[i]).transpose())

# # # flat_data1_datacol = pd.DataFrame(dict(flat_data1_datacol['awarded'])).transpose()
# # flat_data1_datacol.head()
```

# put the normalized data together

```python
fully_flattened_data = pd.concat([data1, data1_datacol, flat_data1_datacol, flattened_col_awarded], axis = 1)
```

```python
pd.set_option("display.max_columns", 0)


fully_flattened_data.head()
```

```python
# since columns 'data', 'purchaser', 'type', 'awarded' are flattened I can drop the duplications

fully_flattened_data= fully_flattened_data.drop(columns=['data', 'purchaser', 'type', 'awarded'])
```
